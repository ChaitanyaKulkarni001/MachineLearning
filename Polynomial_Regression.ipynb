````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
````

````python
dataset = pd.read_csv('/content/drive/MyDrive/Machine Learning-A-Z-Codes-Datasets/Machine Learning A-Z (Codes and Datasets)/Part 2 - Regression/Section 6 - Polynomial Regression/Python/Position_Salaries.csv')
# data.head()

X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
````

# **Training the linear regression model on Whole dataset**

````python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X,y)
````

````python
# # regressor.predict([[1]])
# plt.scatter(X,y,color='red')
# plt.plot(X,regressor.predict(X),color='blue')
# plt.show()
````

# ***Training the Polynomial regression model on Whole dataset***

It is combination of Linear features but it can learn non linear features

````python
from sklearn.preprocessing import PolynomialFeatures

poly_reg = PolynomialFeatures(degree =4)
X_poly = poly_reg.fit_transform(X)

main_regressor = LinearRegression()
main_regressor.fit(X_poly,y)
````

# **Vizualizing the Linear Regression on whole dataset**


````python
plt.scatter(X,y,color='red')
plt.plot(X,regressor.predict(X),color='blue')
plt.title("Truth or Bluff(Linear Regression)")

plt.xlabel("position level")
plt.ylabel("Salary")
````

# **Vizualizing the Polynomial Regression on whole dataset**


````python
plt.scatter(X,y,color='red')
# plt.plot(X,main_regressor.predict(poly_reg.fit_transform(X)),color='blue')
plt.plot(X,main_regressor.predict(X_poly),color='blue')
plt.title("Truth")

plt.xlabel("position level")
plt.ylabel("Salary")
plt.show()
````
