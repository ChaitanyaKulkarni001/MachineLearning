# **Logistic Regression**

image.png

In statistics, maximum likelihood estimation (MLE) is a method of estimating the parameters of an assumed probability distribution, given some observed data. This is achieved by maximizing a likelihood function so that, under the assumed statistical model, the observed data is most probable. The point in the parameter space that maximizes the likelihood function is called the maximum likelihood estimate

# **Importing Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# **Data**

```python
data = pd.read_csv('Social_Network_Ads.csv')
X = data.iloc[:,:-1].values
y = data.iloc[:,-1].values

# data.head()
```

# **Splitting Data into test and training set**

```python
# from sklearn.model_selection import train_test_split
# X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=1)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
print(X_train)
```

# **Feature Scaling**

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)


```

# **Training the model**

```python
from sklearn import linear_model
classifier = linear_model.LogisticRegression()

classifier.fit(X_train,y_train)

# print(classifier)
```

```python
print(X_test)
```

```python
classifier.predict([[-0.29863069,-1.23842019]])
```

```python
print(sc.inverse_transform([[-0.29863069,-1.23842019]]))
```

```python
result = classifier.predict(X_test)
print(np.concatenate((result.reshape(len(result),1), y_test.reshape(len(y_test),1)),1))
# result.reshape(len(result),1)
# print(y_test)
```

# **Making the Confusion Matrix**

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, result)
```

```python
# Accuracy
from sklearn.metrics import accuracy_score
accuracy_score(y_test,result)
```

# **Vizualizing training set**

```python
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(X_train), y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.25))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Logistic Regression (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
